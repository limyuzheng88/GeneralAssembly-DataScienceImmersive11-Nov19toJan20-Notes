```python
import pandas as pd
import psycopg2
```

```python

connection = psycopg2.connect(
    host = 'localhost',
    port = 5432,
    user = 'postgres', #as given in code solution
    password = 'breadpig', #what i had set for my laptop's postgres superuser pswd
    
    #prep: 1.open SQL Shell, type 'enter' ~4x till they ask for pswd, and input as what you had previously set.
    # 2. type: 'CREATE DATABASE titanic'
    # 3. type: '\c titanic', to connect to the 'titanic' database 
    # 4. with connection established, we can access titanic database! we could equally access postgres database.
    # (you could actually see the tables you've created, by opening an SQL IDE eg. dbeaver, tableplus (for 
    # both mac and windows), postico (for mac))
    database='titanic'
    )
```

```python
#creating cursor object
cursor=connection.cursor()
```

```python
# creating 3 tables, called 'passengers' etc, containing columns 'id' (as pri key), 'name' (of
# dtype text) etc respectively

# creating 1st table, with these columns and their datatypes
cursor.execute("""CREATE TABLE passengers(
id SERIAL PRIMARY KEY,
name text,
sex text,
age float,
sibsp integer,
parch integer)""")

# creating 2nd table, with these columns and their datatypes
cursor.execute("""CREATE TABLE survival(
id SERIAL PRIMARY KEY,
survived integer)""")

# creating 3rd table, with these columns and their datatypes
cursor.execute("""CREATE TABLE tripinfo(
id SERIAL PRIMARY KEY,
pclass integer,
ticket text,
fare float,
cabin text,
embarked text)""")

connection.commit() #MUUUST commit, =save. Else changes won't be saved
```

```python
#Quering Information Shema table
sql = """
SELECT "table_name","column_name", "data_type", "table_schema"
FROM INFORMATION_SCHEMA.COLUMNS
WHERE "table_schema" = 'public'
ORDER BY table_name  
"""
pd.read_sql(sql, con=connection)
```

```python
# Copy data from csv files to each of the 3 tables:

with open('passengers.csv', 'r') as row:
    next(row)# Skip the header row.
    cursor.copy_from(row, 'passengers', sep=',')
    
with open('survival.csv', 'r') as row:
    next(row) 
    cursor.copy_from(row, 'survival', sep=',')

with open('trip_info.csv', 'r') as row:
    next(row)
    cursor.copy_from(row, 'tripinfo', sep=',')
    
connection.commit()    
```

```python
# view table 'passengers'
sql = """
SELECT *
FROM passengers
"""

pd.read_sql(sql, con=connection).head()
```

```python
# Create a new Role gastudent
cursor.execute("""CREATE ROLE gastudent WITH LOGIN PASSWORD 'gastudent'""")
#Grant different permissions for a new role
cursor.execute("""GRANT CONNECT ON DATABASE titanic TO gastudent""")
cursor.execute("""GRANT SELECT, UPDATE, INSERT ON passengers TO gastudent""")
cursor.execute("""GRANT SELECT, UPDATE, INSERT ON survival TO gastudent""")
cursor.execute("""GRANT SELECT, UPDATE, INSERT ON tripinfo TO gastudent""")

connection.commit()
```

```python
# simple query to interact with a remote DB.

sql = """
SELECT pclass, AVG(survived)
FROM (SELECT t.pclass, s.survived
FROM tripinfo t
INNER JOIN survival s ON t.id = s.id
) AS x
GROUP BY pclass
"""

pd.read_sql(sql, con=connection)
```
