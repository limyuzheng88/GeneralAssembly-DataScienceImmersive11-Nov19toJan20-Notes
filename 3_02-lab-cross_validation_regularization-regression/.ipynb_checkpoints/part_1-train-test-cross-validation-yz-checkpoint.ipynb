## Review of Train/Test Validation Methods

We've discussed overfitting, underfitting, and how to validate the "generalizeability" of your models by testing them on unseen data. 

In this lab you'll practice two related validation methods: 
1. **train/test split**
2. **k-fold cross-validation**

Train/test split and k-fold cross-validation both serve two useful purposes:
- We prevent overfitting by not using all the data, and
- We retain some remaining data to evaluate our model.

In the case of cross-validation, the model fitting and evaluation is performed multiple times on different train/test splits of the data.

Ultimately we can use the training/test validation framework to compare multiple models on the same dataset. This could be comparisons of two linear models, or of completely different models on the same data.

```python
# define a function, which calculates regression metrics

def reg_metrics(y,prediction): 
    EVRSF = metrics.explained_variance_score(y, prediction) #Explained variance regression score function
    RMSERL = metrics.mean_squared_error(y, prediction)**0.5       #Root Mean squared error regression loss
    R2 = metrics.r2_score(y, prediction)                    #R^2 (coefficient of determination) regression score function.
    print('\nExplained variance regression score function =', EVRSF, 
          '\nMean squared error regression loss =', RMSERL,
          '\nR^2 (coefficient of determination) regression score function =', R2)
    return EVRSF, RMSERL, R2
```

```python
# A:
X = df[['RM','PTRATIO','LSTAT']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)  #split into 50-50 train, test sets
lm = LinearRegression()    #create lin reg model
model = lm.fit(X_train, y_train)    #fir model on training sets
prediction = model.predict(X_test)    #predict on test sets

# scatterplots of var against house median values
f, axes = plt.subplots(1, 3,figsize=(15,3))
sns.scatterplot(X_test['RM'],df_merge['MEDV'],ax=axes[0])
sns.scatterplot(X_test['RM'],prediction,ax=axes[0], alpha=0.5)
sns.scatterplot(X_test['PTRATIO'],df_merge['MEDV'],ax=axes[1])
sns.scatterplot(X_test['PTRATIO'],prediction,ax=axes[1], alpha=0.5)
sns.scatterplot(X_test['LSTAT'],df_merge['MEDV'],ax=axes[2])
sns.scatterplot(X_test['LSTAT'],prediction,ax=axes[2], alpha=0.5)
axes[0].set(ylabel="MEDV (in 1000's)")

EVRSF_50, RMSERL_50, R2_50 = reg_metrics(y_test,prediction)
```

```python
# loop thru all 3 features in X, across all fold numbers, across all combinations of features 

#Importing tqdm for the progress bar
from tqdm import tnrange, tqdm_notebook

#Initialization variables
numb_features, R_squared_list, feature_list, folds = [],[],[],[]

#Looping over k = 1 to k = 3 features in X
for k in tnrange(1,len(X.columns) + 1, desc = 'Loop...'):
    #looping over no. of folds     
    for j in range(2,15):
        #Looping over all possible combinations: from 3 features choose k (1/2/3)
        for combo in itertools.combinations(X.columns,k):
            tmp_result = fit_linear_reg(X[list(combo)],y,j)   #Store temp result 
            R_squared_list.append(tmp_result)
            feature_list.append(combo)
            numb_features.append(len(combo))   
            folds.append(j)   
#Store in DataFrame
R2_scores = pd.DataFrame({'numb_features': numb_features, 'R_squared':R_squared_list,'features':feature_list,'folds':folds})

# print out top R2 results
R2_scores.sort_values(by=['R_squared'], ascending=False).head()
```

Best R2 value of 0.573, from combination of features and folds, as seen abv.
