# Using the requests library

```python
import requests
from bs4 import BeautifulSoup
```

### Status Codes

```python
url = 'https://github.com/apache/spark'
response = requests.get(url)
html = response.text
```

### Creating a `BeautifulSoup` object

```python
soup = BeautifulSoup(html,'lxml')
```

# `soup.find_all()`

Returns a **_LIST_** of soup objects that match your query

```python
soup.find_all('td')
```

```python
all_a = soup.find_all('td',{'class':'content'})
all_a
```

```python
results_list = []

for element in all_a[1:]:    #ignore the 1st result, which isnt a folder
    # start a dictionary to store this item's data
    result = {}
    
    # get the title and full link/url
    a_href = element.find('a')
    if a_href:
        result['title'] = a_href.text   # element text
        result['link'] = a_href['href'] # href link
    results_list.append(result)
```

```python
import pandas as pd
pd.DataFrame(results_list)
```

```python
# lets try another website
url = 'https://www.channelnewsasia.com/'
response = requests.get(url)
html = response.text
print(html)
```

```python
soup = BeautifulSoup(html,'lxml')
all_h3 = soup.find_all('h3',{'class':'teaser__heading'})    #find all h3 tags, with class='teaser__heading' 
all_h3
```

```python
result_list=[]

for element in all_h3:
    result={}
    a_href = element.find('a')
#     print(a_href)
    result['title'] = a_href.text
    result['link'] = a_href['href']
    result_list.append(result)
pd.DataFrame(result_list)
```
