#### 7. Rename Columns
Change the names of the columns to more expressive names so that you can tell the difference the SAT columns and the ACT columns. Your solution should map all column names being changed at once (no repeated singular name-changes). **We will be combining these data with some of the data from 2018, and so you should name columns in an appropriate way**.

**Guidelines**:
- Column names should be all lowercase (you will thank yourself when you start pushing data to SQL later in the course)
- Column names should not contain spaces (underscores will suffice--this allows for using the `df.column_name` method to access columns in addition to `df['column_name']`.
- Column names should be unique and informative (the only feature that we actually share between dataframes is the state).

````python
#code

# create function to rename columns 
def rename_cols(data,prefix_test,suffix_year):
    data=data.rename(columns={'Participation':'Participation %'})
    data.columns = [prefix_test + '-' + str(col) + '-' + suffix_year for col in data.columns] #adds prefix (eg. SAT-, ACT-) to all column headers
    data.columns = map(str.lower, data.columns) #make all headers lowercase
    data.columns = data.columns.str.replace(' ', '_') #replace all spaces with _
    re = ['state' if k==0 else data.columns[k] for k in range(len(data.columns))]
    data.columns = re 
    return data
sat_2017=rename_cols(sat_2017,'sat','2017')
sat_2017.head()
````

#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

- Write a function to calculate standard deviation using the formula above

````python
#code
def standard_dev(data):
    total = 0                                         #initialise total value of (x(i)-miu)^2
    for i in range(len(data)):                        #sum up total value of the above
        total = total + (data[i]-np.mean(data))**2
    std = np.sqrt(total/len(data))
    return std
````

- Use a **dictionary comprehension** to apply your standard deviation function to each numeric column in the dataframe.  **No loops**  
- Assign the output to variable `sd` as a dictionary where: 
    - Each column name is now a key 
    - That standard deviation of the column is the value 
     
*Example Output :* `{'ACT_Math': 120, 'ACT_Reading': 120, ...}`

#### Define a custom function to subplot histograms

We have data for two tests for two years. We only have composite (and not subtest scores) for the 2018 ACT. We should write a function that will take the names of 2+ columns and subplot histograms. While you can use pandas plotting or Seaborn here, matplotlib gives you greater control over all aspects of your plots.

[Helpful Link for Plotting Multiple Figures](https://matplotlib.org/users/pyplot_tutorial.html#working-with-multiple-figures-and-axes)

Here's some starter code:

````python
def subplot_histograms(dataframe, list_of_columns, list_of_titles, list_of_xlabels, list_of_ylabels=''):
    nrows = int(np.ceil(len(list_of_columns)/2)) # Makes sure you have enough rows. Divide by 2, cos 2 diff yrs
    fig, ax = plt.subplots(nrows=nrows, ncols=2) # You'll want to specify your figsize
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    for i, column in enumerate(list_of_columns): # Gives us an index value to get into all our lists
        ax[i].hist(dataframe[column]) # feel free to add more settings
        # Set titles, labels, etc here for each subplot
        ax[i].title.set_text(list_of_titles[i])
        ax[i].set_xlabel(list_of_xlabels)
        ax[i].set_ylabel(list_of_ylabels)
        plt.tight_layout() #space out subplots so they don't overlap
````

#### Statistical Evaluation of Distributions 

**If you feel it's appropriate**, using methods we discussed in class, run hypothesis tests to compare variables of interest in our dataset. 

H0: SAT 2017 and SAT 2018 Total scores were the same, hence equal in difficulty for both years

H1: SAT 2017 and SAT 2018 Total scores were not the same, hence different difficulties for both years

Conduct a 2 tailed t-test, with $\alpha=0.05$

````python
# Code:
print('SAT 2017 mean =',final['sat-total-2017'].mean())
print('SAT 2018 mean =',final['sat-total-2018'].mean())
# Print the difference of the sample means, too.
print('difference =',final['sat-total-2018'].mean()-final['sat-total-2017'].mean())
print()

import scipy.stats
t_stat, p_value = scipy.stats.ttest_ind(final['sat-total-2017'], final['sat-total-2018'], equal_var=False)    #False: uses Welch's distribution, where var is diff for both samples
print('t_stat = '+str(t_stat))
print('p_value = '+str(p_value))
print()

# Generate 500 points on the x axis between -4 and 4:
xpoints = np.linspace(-4,4,500)

# Use `stats.t.pdf` to get values on the probability density function for the t-distribution.
# The second argument is the degrees of freedom: n1 + n2 - 2.
n_sat_total_2017 = len(final['sat-total-2017'])
n_sat_total_2018 = len(final['sat-total-2018'])
ypoints = scipy.stats.t.pdf(xpoints, (n_sat_total_2017+n_sat_total_2018-2), 0, 1)

# Initialize a `matplotlib` "figure."
fig = plt.figure(figsize=(8,5))

# Get the current "axis" out of the figure.
ax = fig.gca()

# Plot the lines using `matplotlib`'s plot function:
ax.plot(xpoints, ypoints, linewidth=3, color='darkred');

# Plot 2 vertical lines for our measured difference in rates' t-statistic.
ax.axvline(t_stat, color='black', linestyle='--', lw=3);
ax.axvline(-t_stat, color='black', linestyle='--', lw=3);

lower_tail = scipy.stats.t.cdf(-abs(t_stat), n_sat_total_2017 + n_sat_total_2018 - 2)
upper_tail = 1 - scipy.stats.t.cdf(abs(t_stat), n_sat_total_2017 + n_sat_total_2018 - 2)
p_value_man = lower_tail + upper_tail

print('p_value =',p_value)
print('p_value_man =',p_value_man)
print()

print('Because our p-value is larger than alpha=0.05, and is statistically insignificant, \
hence H0 cannot be rejected. There is insufficient evidence to say that the difference is non-zero, hence we \
cannot conclude that the difficuties of both tests were different.')
````
