# Scrape images for each product (run this after main notebook '1-CustomerSegmentation_2-RecSys_3-MBA' has been run)

load data_rec dataframe

```python
import pickle

# open a file, where you stored the pickled data
file = open('./webapp/static/pickle_files/data_rec.pkl', 'rb')

# load pickled model
data_rec = pickle.load(file)

# close the file
file.close()

data_rec.head()
```

from data_rec dataframe, scrape image urls for each product. pip install google_images_download if you haven't already

```python
# pip install google_images_download
import random
import numpy as np

# importing google_images_download module 
from google_images_download import google_images_download  
import os

# creating object 
response = google_images_download.googleimagesdownload()  

search_queries = data_rec.Description.unique() 
  
def downloadimages(query):        
    arguments = {"keywords": query, #the search query 
                 "format": "jpg", #image file format
                 "limit":1, #just get 1 image
                 "print_urls":True, #print these for info
                 "print_size":True,
                 "size":'medium', #image size which can be specified manually ("large, medium, icon")
                 "no_download":True, #dont download image, just get urls
                 "socket_timeout":600, #in case of poor internet connection, wait for up to 600s
                 "delay":random.random()} #to scrape without getting my proxy banned
    urls = response.download(arguments) #store the image urls
    return urls

# begin scraping
urls=[]
ct=0
for query in search_queries: 
    urls.append(downloadimages(query))
    ct+=1
    print(ct)
```

store all these urls into a dataframe

```python
Description,url=[],[]
for i in range(len(urls)):
    Description.append(list(urls[i][0].keys())[0])
    try:
        url.append(list(urls[i][0].values())[0][0])
    except:
        url.append('https://ak2.picdn.net/shutterstock/videos/14128382/thumb/11.jpg') #random gift box img for empty urls
urls = pd.DataFrame({'Description':Description,'url':url})

print(len(urls))
urls.head()
```

Now lets merge the urls dataframe, with the product lookup table (from main notebook) and price of each product.

```python
# open a file, where you stored the pickled data
file = open('./webapp/static/pickle_files/lookup.pkl', 'rb')

# load pickled model
lookup = pickle.load(file)

# close the file
file.close()

lookup_urls = pd.merge(lookup,urls,on='Description',how='inner')
# Price may vary for a product, lets just take each product's max price to present on webapp
lookup_urls_price = pd.merge(lookup_urls, data_rec.groupby('StockCode').UnitPrice.max(),on='StockCode') 

# pickle this data
with open('./webapp/static/pickle_files/lookup_urls_price.pkl', 'wb') as file:
    pickle.dump(lookup_urls_price, file)
    
lookup_urls_price.head()
```
